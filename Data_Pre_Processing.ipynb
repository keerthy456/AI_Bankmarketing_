```
from sklearn import preprocessing
```

```
from google.colab import files 
#data = files.upload()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import io
dataset = pd.read_csv('AIphase1.csv')
le = preprocessing.LabelEncoder()
for col in ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome','y']:
    dataset[col] = le.fit_transform(dataset[col])


x = dataset.iloc[:,:-1]
output = dataset.iloc[:,-1] 
size = dataset.shape
#print(dataset.iloc[:5,:])

max_array = []
min_array = []
column= size[1]
ind = 0
names = list(x.columns)
#print("Distribution of Data before Normalization")
"""
for i in range(len(names)):
  print(i+1)
  plt.hist(x.iloc[:,i],bins=15)
  plt.title(names[i])
  plt.show()
"""

"""
NORMALIZATION:
"""
for i in range(column):
  min = dataset.iloc[:,i].min(axis=0)
  max = dataset.iloc[:,i].max(axis=0)
  max_array.append(max)
  min_array.append(min)
  dataset.iloc[:,i] = (dataset.iloc[:,i]-min)/(max-min)

ax = dataset['y'].value_counts().plot(kind='bar')
plt.title("Distribution of Output variable- 'Y'")
plt.xlabel("Output(0/1)")
plt.ylabel("count")
plt.show()

x = dataset.iloc[:,:-1]
print("Distribution of Data after Normalization")
for i in range(len(names)):
  plt.hist(x.iloc[:,i],bins=15)
  plt.title(names[i])
  plt.show()
```
